Nonlinear Regrssion Section 1

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv')
df.describe()
```

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]

plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color='blue')
plt.xlabel('Engine size')
plt.ylabel('CO2 Emissions')
plt.show()
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
from sklearn import preprocessing
from sklearn import linear_model
from sklearn import metrics
x_train = np.asanyarray(train[['ENGINESIZE']])
y_train = np.asanyarray(train[['CO2EMISSIONS']])
x_test = np.asanyarray(test[['ENGINESIZE']])
y_test = np.asanyarray(test[['CO2EMISSIONS']])

poly_transformer = preprocessing.PolynomialFeatures(degree = 2)
x_train_transformed = poly_transformer.fit_transform(x_train)
x_test_transformed = poly_transformer.fit_transform(x_test)

model = linear_model.LinearRegression()
model.fit(x_train_transformed, y_train)
print('Coef: ', model.coef_)
print('Intercept: %.2f' % model.intercept_)
y_hat = model.predict(x_test_transformed)

print('MSE: %.2f' % np.mean((y_hat - y_test) ** 2))
print('MAE: %.2f' % np.mean(np.absolute(y_hat - y_test)))
print('R2-Score: %.2f' % metrics.r2_score(y_test, y_hat))
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')
XX = np.arange(0.0, 9.0, 0.1)
yy = model.intercept_ + model.coef_[0][1] * XX + model.coef_[0][2] * np.power(XX, 2)
plt.plot(XX, yy, '-r')
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
plt.title('Training Data')
```

Nonlinear Regression Section 2

```python
def plot_with_noise(x, y, y_noise):
    y_data = y + y_noise
    plt.plot(x, y_data, 'bo')
    plt.plot(x, y_data, 'r')
    plt.xlabel('Dep Var')
    plt.ylabel('Indep Var')
    plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = 2 * (x) + 3
y_noise = np.random.normal(size=x.size)

plot_with_noise(x, y, y_noise)
```

```python
y = 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 20 * np.random.normal(size=x.size)

plot_with_noise(x, y, y_noise)
```

```python
y = 3 * np.power(2, x) + 1
plot_with_noise(x, y, 0)
```

```python
x = np.arange(0.1, 10.0, 0.1)
y = np.log(x)

plot_with_noise(x, y, 0)
```

Logistic

```python
x = np.arange(-5.0, 5.0, 0.1)
y = 1 - 4/(1 + np.power(3, x))

plot_with_noise(x, y, 0)
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/china_gdp.csv')

plt.figure(figsize=(8, 5))
x_data, y_data = (df['Year'].to_numpy(), df['Value'].to_numpy())
plt.plot(x_data, y_data, 'ro')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
Beta_1 = -1
Beta_2 = 1
y = 1/(1 + np.exp(Beta_1 * (x - Beta_2)))

plot_with_noise(x, y, 0)
```

```python
def sigmoid(x, Beta_1, Beta_2):
    y = 1/(1 + np.exp(-Beta_1 * (x - Beta_2)))
    return y
```

```python
beta_1 = 0.1
beta_2 = 1990.0

y_pred = sigmoid(x_data, beta_1, beta_2)
norm_const = 15000000000000.
plt.plot(x_data, y_pred * norm_const)
plt.plot(x_data, y_data, 'ro')
```

```python
x_norm = x_data/max(x_data)
y_norm = y_data/max(y_data)
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, x_norm, y_norm)
print('beta_1 = %f \nbeta_2 = %f' % (popt[0], popt[1]))
```

```python
x_test = np.linspace(1960, 2015, 55)
x_test = x_test/max(x_test)
plt.figure(figsize=(8,5))
y_hat = sigmoid(x_test, *popt)
"""
    The test data and training data are the same
"""
plt.plot(x_norm, y_norm, 'ro', label = 'train')
plt.plot(x_test, y_hat, linewidth = 3.0, label = 'test')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
print('MSE: %f' % np.mean((y_hat - y_norm) ** 2))
print('MAE: %f' % np.mean(np.absolute(y_hat - y_norm)))
print('R2-Score: %f' % metrics.r2_score(y_norm, y_hat))
```
