```python
import torch
import matplotlib.pyplot as plt
import numpy as np
torch.manual_seed(0)
```

```python
def show_data(data_sample, shape=(28, 28)):
    plt.imshow(data_sample[0].numpy().reshape(shape), cmap='gray')
    plt.title(data_sample[1])
    plt.show()
```

```python
from torchvision import transforms
from torchvision import datasets
```

```python
ds = datasets.MNIST(root='./data/',
                   train = False,
                   download=True,
                   transform = transforms.ToTensor()
)
```

```python
# Plot the first element in the dataset
show_data(ds[0])
```

```python
show_data(ds[1])
```

```python
# Combine two transforms: crop and convert to tensor. Apply the compose to MNIST dataset
crop_tensor_transform = transforms.Compose([transforms.CenterCrop(20), transforms.ToTensor()])
ds = datasets.MNIST(root='./data/',
                   train = False,
                   download=True,
                   transform = crop_tensor_transform)
```

```python
# Show first two elements
show_data(ds[0], shape=(20, 20))
```

```python
show_data(ds[1], shape=(20, 20))
```

```python
fliptensor_transform = transforms.Compose([transforms.RandomVerticalFlip(p=1), transforms.ToTensor()])
ds = datasets.MNIST(root='./data/',
                   train = False,
                   download=True,
                   transform = fliptensor_transform)
```

```python
# Show second element
show_data(ds[1])
```

```python
# Practice: Combine vertical flip, horizontal flip and convert to tensor as a compose. Apply the compose on image. Then plot the image
custom = transforms.Compose([transforms.RandomVerticalFlip(p=1), transforms.RandomHorizontalFlip(p=1), transforms.ToTensor()])
ds = datasets.MNIST(root='./data/',
                   train = False,
                   download=True,
                   transform = custom)
show_data(ds[1])
```
