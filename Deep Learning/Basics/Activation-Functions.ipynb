```python
import torch
from torch import nn, optim
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
torch.manual_seed(2)
```

```python
z = torch.arange(-10, 10, 0.1).view(-1, 1)
sig = nn.Sigmoid()
y_sig = sig(z)
```

```python
plt.plot(z.numpy(), y_sig.detach().numpy())
plt.show()
```

```python
Tanh = nn.Tanh()
y_tanh = Tanh(z)
plt.plot(z.numpy(), y_tanh.detach().numpy())
plt.show()
```

```python
ReLU = nn.ReLU()
y_relu = ReLU(z)
plt.plot(z.numpy(), y_relu.detach().numpy())
plt.show()
```

```python
plt.plot(z.numpy(), y_sig.detach().numpy(), label='Sigmoid')
plt.plot(z.numpy(), y_tanh.detach().numpy(), label='Tanh')
plt.plot(z.numpy()[0:125], y_relu.detach().numpy()[0:125], label='ReLU')
plt.show()
```

```python
# Compare the activation functions again using a tensor in the range (-1, 1)
x = torch.arange(-1, 1, 0.1)
y_sig = sig(x)
y_tanh = Tanh(x)
y_relu = ReLU(x)

plt.plot(x.numpy(), y_sig.detach().numpy())
plt.plot(x.numpy(), y_tanh.detach().numpy())
plt.plot(x.numpy(), y_relu.detach().numpy())
plt.plot()
```
