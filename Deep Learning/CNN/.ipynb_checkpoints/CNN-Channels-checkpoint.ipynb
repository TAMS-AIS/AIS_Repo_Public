```python
import torch
from torch import nn
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
torch.manual_seed(0)
```

```python
def show(data, shape):
    plt.imshow(data.reshape(shape), cmap='gray')
    plt.colorbar()
    plt.show()
```

```python
# Multiple out channels
conv1 = nn.Conv2d(in_channels=1, out_channels=3, kernel_size=(3, 3))
```

```python
x = torch.FloatTensor([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
y = torch.FloatTensor([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

conv1.state_dict()['weight'][0][0] = x
conv1.state_dict()['weight'][1][0] = y
conv1.state_dict()['weight'][2][0] = torch.ones(3, 3)
conv1.state_dict()['bias'][:] = torch.FloatTensor([0, 0, 0])
conv1.state_dict()
```

```python
image = torch.zeros(1, 1, 5, 5)
image[0, 0, :, 2] = 1
show(image.numpy(), (5, 5))
```

```python
z = conv1(image)
print(z)
```

```python
#Multipple in and out channels
conv3 = nn.Conv2d(2, 3, (3, 3))

conv3.state_dict()['weight'][0][:] = x * 0.5
conv3.state_dict()['weight'][1][0] = x
conv3.state_dict()['weight'][1][1] = x * -1
conv3.state_dict()['weight'][2][0] = torch.tensor([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
conv3.state_dict()['weight'][2][1] = torch.tensor([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
conv3.state_dict()['bias'][:] = torch.tensor([0, 0, 0])

conv3.state_dict()
```

```python
z = conv3(image3)
for i, channel in enumerate(z[0]):
    plt.title(f'Channel {i}')
    show(channel.detach().numpy(), (3, 3))
```
