```python
import torch
from torchvision import datasets, transforms
from torch import optim, nn
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
torch.manual_seed(0)
```

```python
def plot_channels(W):
    n_out = W.shape[0]
    n_in = W.shape[1]
    w_min = W.min().item()
    w_max = W.max().item()
    fig, axes = plt.subplots(n_out, n_in)
    fig.subplots_adjust(hspace=0.1)
    out_index = 0
    in_index = 0
    
    #plot outputs as rows inputs as columns 
    for ax in axes.flat:
        if in_index > n_in-1:
            out_index = out_index + 1
            in_index = 0
        ax.imshow(W[out_index, in_index, :, :], vmin=w_min, vmax=w_max, cmap='seismic')
        ax.set_yticklabels([])
        ax.set_xticklabels([])
        in_index = in_index + 1

    plt.show()
```

```python
def plot_parameters(W, number_rows=1, name="", i=0):
    W = W.data[:, i, :, :]
    n_filters = W.shape[0]
    w_min = W.min().item()
    w_max = W.max().item()
    fig, axes = plt.subplots(number_rows, n_filters // number_rows)
    fig.subplots_adjust(hspace=0.4)

    for i, ax in enumerate(axes.flat):
        if i < n_filters:
            # Plot the image.
            ax.imshow(W[i, :], vmin=w_min, vmax=w_max, cmap='seismic')
            ax.set_xticks([])
            ax.set_yticks([])
    plt.suptitle(name, fontsize=10)    
    plt.show()
```

```python
def show(data, y, shape):
    plt.imshow(data.reshape(shape), cmap = 'gray')
    labels = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
    plt.title('label = ' + str(labels[y]))
    plt.colorbar()
    plt.show()
```

```python
train_ds = datasets.FashionMNIST(root='../data', download=True, train=True, transform=transforms.ToTensor())
val_ds = datasets.FashionMNIST(root='../data', download=True, train=False, transform=transforms.ToTensor())
```

```python
show(train_ds[0][0].numpy(), train_ds[0][1], (28, 28))
```

```python
class CNN(nn.Module):
    def __init__(self, out_1=28, out_2=56):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1, out_1, 5, padding=2) # Padding of 2 to make whole kernel go through 30 by 30 area
        self.maxpool1 = nn.MaxPool2d(2)
        self.conv2 = nn.Conv2d(out_1, out_2, 5, padding=2)
        self.maxpool2 = nn.MaxPool2d(2)
        self.lin1 = nn.Linear(out_2 * 7 * 7, 10) # Output of maxpooling will be 7 by 7
        
    def forward(self, x):
        x = torch.relu(self.conv1(x))
        x = self.maxpool1(x)
        x = torch.relu(self.conv2(x))
        x = self.maxpool2(x)
        x = self.lin1(x.view(x.size(0), -1))
        return x
```

```python
model = CNN()
```

```python
plot_parameters(model.state_dict()['conv1.weight'], number_rows=4, name="1st layer kernels before training ")
plot_parameters(model.state_dict()['conv2.weight'], number_rows=4, name='2nd layer kernels before training' )
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)
train_loader = DataLoader(train_ds, batch_size=30)
val_loader = DataLoader(val_ds, batch_size=500)
```

```python
def train_model(model, optimizer, train_loader, val_loader, criterion, epochs):
    results = {'training loss': [], 'val accuracy': []}
    for epoch in range(epochs):
        epoch_cost = 0
        for x, y in train_loader:
            y_hat = model(x)
            loss = criterion(y_hat, y)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            epoch_cost += loss.item()
        results['training loss'].append(epoch_cost)
        
        correct = 0
        for x, y in val_loader:
            y_hat = model(x)
            _, label = torch.max(y_hat.data, 1)
            correct += (y == label).sum().item()
        results['val accuracy'].append(correct / len(val_loader.dataset))
        print(epoch, end=' ')
    return results
```

```python
train_model(model, optimizer, train_loader, val_loader, criterion, 5)
```

```python
plot_parameters(model.state_dict()['conv1.weight'], number_rows=4, name="1st layer kernels after training ")
plot_parameters(model.state_dict()['conv2.weight'], number_rows=4, name='2nd layer kernels after training' )
```

```python
# Show 5 incorrect predictions on val data
i = 0
for x, y in DataLoader(val_ds, batch_size=1):
    y_hat = model(x)
    _, label = torch.max(y_hat, 1)
    if label != y:
        labels = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
        print('pred =', labels[label.detach().numpy().item()])
        show(x.numpy(), int(y.numpy()), (28, 28))
        i += 1
    if i > 5:
        break
```
