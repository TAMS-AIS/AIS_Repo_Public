```python
import torch
from torch import nn, optim
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage, misc
%matplotlib inline
torch.manual_seed(0)
```

```python
def show(data, shape):
    plt.imshow(data.reshape(shape), cmap='gray')
    plt.show()
```

```python
conv = nn.Conv2d(1, 1, (3, 3))
x = torch.FloatTensor([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
show(x.numpy(), (3, 3))
```

```python
conv.state_dict()['weight'][0] = x
conv.state_dict()['bias'] = 0
conv.state_dict()
```

```python
image = torch.zeros(1, 1, 5, 5)
image[0,0, :, 2] = 1
show(image, (5, 5))
```

```python
z = conv(image)
print(z)
show(z.detach().numpy(), (3, 3))
```

```python
A = torch.relu(z)
show(A.detach().numpy(), (3, 3))
```

```python
image1 = torch.tensor([[ 1.,  2.,  3., -4.], [ 0.,  2., -3.,  0.], [ 0.,  2.,  3.,  1.], [ 0.,  0.,  0.,  0.]]).view(1, 1, 4, 4)
show(image1, (4, 4))
```

```python
max1 = nn.MaxPool2d(kernel_size=(2, 2), stride=1)
show(max1(image1), (3, 3))
```

```python
# Setting the stride to None (default) will make it stride to cover whole image
max2 = nn.MaxPool2d((2, 2))
show(max2(image1), (2, 2))
```
