## Dropout for Classification task

```python
import torch
from torch import nn, optim
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import DataLoader, Dataset
import torch.nn.functional as F 
%matplotlib inline
```

```python
def accuracy(model, ds):
    _, y_hat = torch.max(model(ds.x), 1)
    return (y_hat == ds.y).numpy().mean()
```

```python
class Data(Dataset):
    def __init__(self, n_samples=1000, noise_std=0.15, train=True):
        coef = np.matrix([-1, 1, 2, 1, 1, -3, 1]).T
        self.x = np.matrix(np.random.randn(n_samples, 2))
        self.f = np.array(coef[0] + self.x * coef[1:3] + np.multiply(self.x[:, 0], self.x[:, 1]) * coef[4] + np.multiply(self.x, self.x) * coef[5:7]).flatten()
        self.coef = coef
        
        self.y = np.zeros(n_samples)
        self.y[self.f > 0] = 1
        self.y = torch.from_numpy(self.y).type(torch.LongTensor)
        self.x = torch.from_numpy(self.x).type(torch.FloatTensor)
        self.x = self.x + noise_std * torch.randn(self.x.size())
        self.f = torch.from_numpy(self.f)
        if train:
            torch.manual_seed(1)
            self.x = self.x + noise_std * torch.rand(self.x.size()) # More noise for training data (for exposing overfitting)
            torch.manual_seed(0)
        self.len = self.x.shape[0]
        
    def __getitem__(self, index):
        return self.x[index], self.y[index]
    
    def __len__(self):
        return self.len
    
    def plot_data(self):
        x = self.x.numpy()
        y = self.y.numpy()
        x_min, x_max = x[:, 0].min(), x[:, 0].max()
        y_min, y_max = x[:, 1].min(), x[:, 1].max()
        
        x_mesh, y_mesh = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
        z = self.multi_dim_poly(np.c_[x_mesh.ravel(), y_mesh.ravel()]).flatten()
        f = np.zeros(z.shape)
        f[z > 0] = 1
        f = f.reshape(x_mesh.shape)
        
        plt.title('True decision boundary  and sample points with noise ')
        plt.plot(self.x[self.y == 0, 0].numpy(), self.x[self.y == 0,1].numpy(), 'bo', label='y=0') 
        plt.plot(self.x[self.y == 1, 0].numpy(), self.x[self.y == 1,1].numpy(), 'ro', label='y=1')
        plt.contour(x_mesh, y_mesh, f,cmap=plt.cm.Paired)
        plt.xlim(0,1)
        plt.ylim(0,1)
        plt.legend()
        
    def multi_dim_poly(self, x):
        x = np.matrix(x)
        out = np.array(self.coef[0] + x * self.coef[1:3] + np.multiply(x[:, 0], x[:, 1]) * self.coef[4] + np.multiply(x, x) * self.coef[5:7])
        return out
```

```python
train_data = Data(3000, 0.2)
train_data.plot_data()
```

```python
torch.manual_seed(0)
val_data = Data(train=False)
```

```python
class Net(nn.Module):
    def __init__(self, dim_in, dim_H, dim_out, drop_p, act_func):
        super(Net, self).__init__()
        self.drop = nn.Dropout(drop_p)
        self.Lin1 = nn.Linear(dim_in, dim_H)
        self.Lin2 = nn.Linear(dim_H, dim_H)
        self.Lin3 = nn.Linear(dim_H, dim_out)
        self.act_func = act_func
        
    def forward(self, x):
        x = self.drop(self.Lin1(x))
        x = self.act_func(x)
        x = self.drop(self.Lin2(x))
        x = self.act_func(x)
        x = self.Lin3(x)
        return x
```

```python
model_nodrop = Net(2, 300, 2, 0, F.relu)
model_drop = Net(2, 300, 2, 0.5, F.relu)
```

```python
model_drop.train()
```

```python
opt_nodrop = optim.Adam(model_nodrop.parameters(), lr=0.01)
opt_drop = optim.Adam(model_drop.parameters(), lr=0.01)
criterion = nn.CrossEntropyLoss()
```

```python
Loss = {'train no drop': [], 'train drop': [], 'val no drop': [], 'val drop': []}
```

```python
def train_models(epochs):
    for epoch in range(epochs):
        # Full batch
        model_drop.train()
        y_hat = model_nodrop(train_data.x)
        y_hat_drop = model_drop(train_data.x)
        loss = criterion(y_hat, train_data.y)
        loss_drop = criterion(y_hat_drop, train_data.y)
        loss.backward()
        loss_drop.backward()
        opt_nodrop.step()
        opt_drop.step()
        opt_nodrop.zero_grad()
        opt_drop.zero_grad()
        
        model_drop.eval()
        y_pred = model_nodrop(val_data.x)
        y_pred_drop = model_drop(val_data.x)
        val_loss = criterion(y_pred, val_data.y)
        val_loss_drop = criterion(y_pred_drop, val_data.y)
        
        Loss['train no drop'].append(loss.item())
        Loss['train drop'].append(loss_drop.item())
        Loss['val no drop'].append(val_loss.item())
        Loss['val drop'].append(val_loss_drop.item())
        
        
        if epoch % 50 == 0:
            print(epoch)
```

```python
train_models(500)
```

```python
plt.figure(figsize=(6.1, 10))
for key, value in Loss.items():
    plt.plot(np.log(np.array(value)), label=key)
plt.xlabel('Epoch')
plt.ylabel('Log of loss') # Not to be confused with Log Loss
plt.legend()
plt.show()
```
