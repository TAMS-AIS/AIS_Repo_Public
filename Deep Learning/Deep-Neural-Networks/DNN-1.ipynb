```python
import torch
from torch import nn, optim
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms, datasets
torch.manual_seed(2)
```

```python
class Net(nn.Module):
    def __init__(self, dim_in, dim_H1, dim_H2, dim_out, act_func):
        super(Net, self).__init__()
        self.Seq1 = nn.Sequential(nn.Linear(dim_in, dim_H1), act_func)
        self.Seq2 = nn.Sequential(nn.Linear(dim_H1, dim_H2), act_func)
        self.Seq3 = nn.Linear(dim_H2, dim_out)
        
    def forward(self, x):
        x = self.Seq1(x)
        x = self.Seq2(x)
        x = self.Seq3(x)
        return x
```

```python
def train_model(model, criterion, trainloader, val_loader, optimizer, epochs):
    results_dict = {'training loss': [], 'validation accuracy': []}
    for epoch in range(epochs):
        if epoch % 5 == 0:
            print(epoch) # Used to track progress
        epoch_loss = 0
        for x, y in trainloader:
            y_hat = model(x.view(-1, 28 * 28))
            loss = criterion(y_hat, y)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            epoch_loss += loss.item()
        results_dict['training loss'].append(epoch_loss)
    
        correct = 0
        for x, y in val_loader.dataset:
            y_hat = model(x.view(-1, 28 * 28))
            _, label = torch.max(y_hat, 1)
            correct += (label == y).sum().item()
        accuracy = correct / (len(val_loader.dataset))
        results_dict['validation accuracy'].append(accuracy)
    return results_dict
```

```python
train_data = datasets.MNIST(root='../data', train=True, download=True, transform=transforms.ToTensor())
val_data = datasets.MNIST(root='../data', train=False, download=True, transform=transforms.ToTensor())

from torch.utils.data import DataLoader
trainloader = DataLoader(train_data, batch_size=30, shuffle=True)
val_loader = DataLoader(val_data, batch_size=500)

criterion = nn.CrossEntropyLoss()
```

```python
model_sig = Net(28 * 28, 50, 50, 10, nn.Sigmoid()) # Instead of 100 neurons in single H Layer, 50 in each of two H Layers
optim_sig = optim.SGD(model_sig.parameters(), lr=0.01)

model_tanh = Net(28 * 28, 50, 50, 10, nn.Sigmoid())
optim_tanh = optim.SGD(model_tanh.parameters(), lr=0.01)

model_relu = Net(28 * 28, 50, 50, 10, nn.ReLU())
optim_relu = optim.SGD(model_relu.parameters(), lr=0.01)
```

```python
results_sig = train_model(model_sig, criterion, trainloader, val_loader, optim_sig, 30)
```

```python
results_tanh = train_model(model_tanh, criterion, trainloader, val_loader, optim_tanh, 30)
```

```python
results_relu = train_model(model_relu, criterion, trainloader, val_loader, optim_relu, 30)
```

```python
plt.plot(results_sig['training loss'], label='Sigmoid')
plt.plot(results_tanh['training loss'], label='Tanh')
plt.plot(results_relu['training loss'], label='ReLU')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend()
plt.show()
```

```python
plt.plot(results_sig['validation accuracy'], label='Sigmoid')
plt.plot(results_tanh['validation accuracy'], label='Tanh')
plt.plot(results_relu['validation accuracy'], label='ReLU')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.show()
```
