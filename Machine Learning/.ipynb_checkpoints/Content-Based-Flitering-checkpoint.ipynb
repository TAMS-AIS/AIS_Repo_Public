```python
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
import numpy as np
%matplotlib inline
```

```python
movies_df = pd.read_csv('data/ml-latest/movies.csv')
movies_df.head()
```

```python
# ratings_df = pd.read_csv('data/ml-latest/ratings.csv')
# rantings_df.head() # Could not do this lab because of JAVA runtime error loading the csv file
```

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))', expand=False)
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)', expand=False)
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '')
movies_df.head()
```

```python
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

```python
movies_matrix_df = movies_df.copy()
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        movies_matrix_df.at[index, genre] = 1
movies_matrix_df = movies_matrix_df.fillna(0)
movies_matrix_df.head()
```

```python
user_titles = movies_df['title'][np.random.randint(low = 0, high = 30000, size=(5,))]
test = pd.DataFrame(user_titles).reset_index()
```

```python
userInput = [
    {'title': 'Breakfast Club, The', 'rating': 5},
    {'title':'Toy Story', 'rating':3.5},
    {'title':'Jumanji', 'rating':2},
    {'title':'Pulp Fiction', 'rating':5},
    {'title':'Akira', 'rating':4.5}
]
temp = pd.DataFrame(userInput)
user_df = pd.DataFrame({'title': test['title'], 'ratings': temp['rating']})
user_df
```

```python
# print(movies_df['title'].isin(user_df['title'].tolist()).value_counts())
inputId = movies_df[movies_df['title'].isin(user_df['title'].tolist())]
user_df = pd.merge(inputId, user_df)
user_df = user_df.drop('genres', 1).drop('year', 1)
user_df
```

```python
user_movies = movies_matrix_df[movies_matrix_df['movieId'].isin(user_df['movieId'].tolist())]
user_movies
```

```python
user_movies = user_movies.reset_index(drop=True)
user_genre_matrix = user_movies.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
user_genre_matrix
```

```python
# print(user_genre_matrix.transpose(), user_df['ratings'])
user_profile = user_genre_matrix.transpose().dot(user_df['ratings'])
user_profile
```

```python
genre_table = movies_matrix_df.set_index(movies_matrix_df['movieId'])
genre_table = genre_table.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
genre_table.head()
```

```python
#Multiply the genres by the weights and then take the weighted average
reco_df = ((genre_table * user_profile).sum(axis=1)/(user_profile.sum()))
# reco_df.head()
reco_df = reco_df.sort_values(ascending=False)
reco_df = reco_df.head(20)
reco_df
```

```python
# temp = pd.
reco_1 = movies_df[movies_df['movieId'].isin(reco_df.head(20).keys())]
reco_1 = reco_1.reset_index(drop=False)
reco_1
```

```python
final_reco_df = pd.DataFrame({'movieId': reco_1['movieId'],
                     'title': reco_1['title'],
                     'genres': reco_1['genres'],
                     'year': reco_1['year'],
                     'conf_val': reco_df.reset_index(drop=True)})
final_reco_df
```
