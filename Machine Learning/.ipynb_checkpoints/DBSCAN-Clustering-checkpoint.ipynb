```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
def createDataPoints(centroidLocation, numSamples, clusterDeviation):
    x, y = make_blobs(n_samples=numSamples, centers=centroidLocation, cluster_std=clusterDeviation)
    x = StandardScaler().fit_transform(x)
    return x, y
```

```python
x, y = createDataPoints([[4,3], [2,-1], [-1,4]], 1500, 0.5)

epsilon = 0.3
min_sam = 7
model = DBSCAN(eps=epsilon, min_samples=min_sam)
model.fit(x)
labels = model.labels_
print(labels)
print(np.unique(labels))
```

```python
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[model.core_sample_indices_] = True
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(n_clusters)
```

```python
unique_labels = set(labels)
unique_labels
```

```python
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

for k, col in zip(unique_labels, colors):
    if k == -1:
        col = 'k' # for outliers
    class_member_mask = (labels == k)
    xy = x[class_member_mask & core_samples_mask]
    plt.scatter(xy[:,0], xy[:,1], s=50, c=[col], marker='o', alpha=0.5)
    xy = x[class_member_mask & ~core_samples_mask]
    plt.scatter(xy[:,0], xy[:,1], s=50, c=[col], marker='o', alpha=0.5)
    plt.scatter
```

```python
# write your code here
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(x)
labels = model.labels_
centroids = model.cluster_centers_
```

```python
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(labels))))
for k, col in zip(set(labels), colors):
    x_curr = (labels == k)
    plt.scatter(x[x_curr,0], x[x_curr,1], s=50, marker='.', c=[col], alpha=0.1)
    plt.scatter(centroids[k,0], centroids[k,1], marker='o', c=[col], alpha=1)
```
