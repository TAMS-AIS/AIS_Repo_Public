```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
from sklearn import manifold, datasets
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
x, y = make_blobs(n_samples=50, centers=[[4,4], [-2, -1], [1, 1], [10,4]], cluster_std=0.9)
plt.scatter(x[:, 0], x[:, 1], marker='o')
```

```python
model = AgglomerativeClustering(n_clusters=4, linkage='average')
model.fit(x, y)
```

```python
plt.figure(figsize=(6,4))
x_min, x_max = np.min(x, axis=0), np.max(x, axis=0)
x = (x - x_min)/(x_max - x_min)

for i in range(x.shape[0]):
    plt.text(x[i, 0], x[i, 1], str(y[i]),
            color=plt.cm.nipy_spectral(model.labels_[i]/10.),
            fontdict={'weight': 'bold', 'size': 9})

plt.xticks([])
plt.yticks([])
plt.scatter(x[:, 0], x[:, 1])
plt.title('Predicted clusters')
plt.show()
```

```python
dist_matrix = distance_matrix(x, x)
print(dist_matrix)
```

```python
z = hierarchy.linkage(dist_matrix, 'complete')
```

```python
dendro = hierarchy.dendrogram(z)
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%204/data/cars_clus.csv')
df.head()
```

```python
# Data cleaning
print('shape:', df.shape)
df[['sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales']] = df[['sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales']].apply(pd.to_numeric, errors='coerce')
df = df.dropna()
df = df.reset_index(drop=True)
pdf = df
print('Shape after cleaning:', df.shape)
```

```python
# Feature Selection
df = df[['engine_s',  'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap', 'mpg']]
```

```python
from sklearn.preprocessing import MinMaxScaler
x = df.values
min_max = MinMaxScaler()
scaled_arr = min_max.fit_transform(x)
print(scaled_arr[:5,])
```

```python
# Clustering with scipy
import scipy
leng = scaled_arr.shape[0]
D = np.zeros([leng, leng])
for i in range(leng):
    for j in range(leng):
        D[i, j] = scipy.spatial.distance.euclidean(scaled_arr[i], scaled_arr[j])
print(D)
```

```python
import pylab as pl
z = hierarchy.linkage(D, 'complete')
dendro = hierarchy.dendrogram(z)
```

```python
clusters = hierarchy.fcluster(z, 3, criterion='distance')
clusters
```

```python
clusters = hierarchy.fcluster(z, 5, criterion='maxclust')
clusters
```

```python
dist_matrix = distance_matrix(scaled_arr, scaled_arr)
print(dist_matrix)
```

```python
model = AgglomerativeClustering(n_clusters=6, linkage='complete')
model.fit(scaled_arr)
model.labels_
```

```python
pdf['cluster'] = model.labels_
pdf.head()
```

```python
import matplotlib.cm as cm
n_clusters = max(model.labels_) + 1
colors = cm.rainbow(np.linspace(0, 1, n_clusters))
cluster_labels = list(range(0, n_clusters))

plt.figure(figsize=(16, 14))
for color, label in zip(colors, cluster_labels):
    subset = pdf[pdf['cluster'] == label]
    for i in subset.index:
        plt.text(subset.horsepow[i], subset.mpg[i], str(subset['model'][i]), rotation=25)
    plt.scatter(subset.horsepow, subset.mpg, s=subset.price*10, c=[color,], label='cluster' + str(label), alpha=0.5)
    
plt.legend()
plt.title('Clusters')
plt.xlabel('Horsepower')
plt.ylabel('mpg')
```

```python
pdf.groupby(['cluster', 'type'])['cluster'].count()
```

```python
agg_cars = pdf.groupby(['cluster', 'type'])[['horsepow', 'engine_s', 'mpg', 'price']].mean()
print(agg_cars)
```

```python
plt.figure(figsize=(16,10))
for color, label in zip(colors, cluster_labels):
    subset = agg_cars.loc[(label,),]
    for i in subset.index:
        plt.text(subset.loc[i][0] + 5, subset.loc[i][2], 'type='+str(int(i)) + ', price='+str(int(subset.loc[i][3])) + 'k')
    plt.scatter(subset.horsepow, subset.mpg, s=subset.price*20, c=[color,], label='cluster'+str(label))
plt.legend()
plt.title('Clusters')
plt.xlabel('horsepow')
plt.ylabel('mpg')    
```
