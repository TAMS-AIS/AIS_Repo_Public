k-Means Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
%matplotlib inline
```

```python
np.random.seed(0)
def_centers = [[4, 4], [-2, -1], [2, -3], [1, 1]]
x_data, y_data = make_blobs(n_samples=5000, centers=[[4, 4], [-2, -1], [2, -3], [1, 1]], cluster_std=0.9)
plt.scatter(x_data[:, 0], x_data[:, 1], marker='.')
```

```python
model = KMeans(n_clusters=4, n_init=12)
model.fit(x_data)
labels = model.labels_ # Labels are the prediction of the kmeans
centers = model.cluster_centers_
```

```python
fig = plt.figure(figsize=(6, 4))
colors = plt.cm.Spectral(np.linspace(0, 1, len(np.unique(labels))))
ax = fig.add_subplot(1, 1, 1)

for k, col in zip(range(len(def_centers)), colors):
    members = (labels == k) # Points are labeled true or false whether they belong in a cluster
    cluster_center = centers[k]
    ax.plot(x_data[members, 0], x_data[members, 1], 'w', markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)
    
ax.set_title('KMeans')
ax.set_xticks(())
ax.set_yticks(())
plt.show()
```

KMeans Section 2 with real data

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%204/data/Cust_Segmentation.csv')
df
```

```python
new_df = df.drop('Address', axis=1) # [DONE]
from sklearn.preprocessing import StandardScaler
x_data = new_df.drop('Defaulted', axis=1)
x_data = np.nan_to_num(x_data)
x_data = StandardScaler().fit_transform(x_data)
```

```python
model = KMeans(n_clusters=3, n_init=12)
model.fit(x_data)
labels = model.labels_
centers = model.cluster_centers_
```

```python
new_df['Cluster'] = labels

area = np.pi * (x_data[:, 1] ** 2)
plt.scatter(x_data[:,0], x_data[:, 3], s=area, c=labels.astype(np.float), alpha=0.5)
plt.xlabel('Age', fontsize=18)
plt.ylabel('Income', fontsize=16)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,6))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)

plt.cla()
ax.set_xlabel('Education')
ax.set_ylabel('Age')
ax.set_zlabel('Income')
ax.scatter(x_data[:,1], x_data[:,0], x_data[:,3], c=labels.astype(np.float))
```
